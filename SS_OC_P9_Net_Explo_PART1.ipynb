### Rappel de l'équation:  disponibilité intérieure = Production + Importations - Exportations + Variation de stock
### soit : Exportations = Production + Importations + Variation de stock - disponibilité intérieure

```python
# Remplacer les NaN par la valeur calculée dans 'Exportation'
def calcul_exportations(row):
    if pd.isna(row['Exportations - Quantité / Milliers de tonnes']):
        return (row['Production / Milliers de tonnes'] +
                row['Importations - Quantité / Milliers de tonnes'] +
                row['Variation de stock / Milliers de tonnes'] -
                row['Disponibilité intérieure / Milliers de tonnes'])
    else:
        return row['Exportations - Quantité / Milliers de tonnes']

# Appliquer la formule dans "Exportations - Quantité / Milliers de tonnes"
Table_Volaille['Exportations - Quantité / Milliers de tonnes'] = Table_Volaille.apply(calcul_exportations, axis=1)
Table_Volaille.isnull().sum()
```

```python
# on peut maintenant supprimer les variations de stocks
Table_Volaille=Table_Volaille.drop('Variation de stock / Milliers de tonnes', axis=1)
```
